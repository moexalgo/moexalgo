### Marketdata

https://moexalgo.github.io/des/realtime/

```python
stocks = Market('EQ') # FUT - фьючерсы, FX - валюта 

# список акций на сегодня
stocks.tickers().head()
```

```python
# торговая статистика за сегодня
stocks.marketdata().head()
```

```python
sber = Ticker('SBER')

# свечки, periods: '1min', '10min', '1h', '1d', '1w', '1m'
sber.candles(start='2025-01-01', end='2025-01-30', period='1d').head()
```

```python
# стакан котировок
sber.orderbook().head()
```

```python
# сделки за сегодня
sber.trades().head()
```

### SuperCandles

https://moexalgo.github.io/des/supercandles/

```python
# статистика по сделкам
sber.tradestats(start='2025-01-25', end='2025-01-30').head()
```

```python
# статистика по заявкам
sber.orderstats(start='2025-01-25', end='2025-01-30').head()
```

```python
# статистика по стакану
sber.obstats(start='2025-01-25', end='2025-01-30').head()
```

###  HI2

```python
sber.hi2(start='2025-01-01', end='2025-03-01').head()
```
