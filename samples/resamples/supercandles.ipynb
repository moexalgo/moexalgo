```python
import os

import moexalgo
from moexalgo import session
from moexalgo.beta.resample import resample

session.TOKEN = os.environ['APIKEY']
```

В **экспериментальном порядке** реализована функция ресемплирования суперсвечей, ендпоинты: `tradestats`, `orderstats`, `obstats`.
Доступны следующие таймфреймы: '5min', '10min', '15min', '20min', '30min', '1h', '2h', '3h', '6h', '12h', '1D', '5D', '10D', '1W', '2W', '4W', '1M'.

```python
sber = moexalgo.Ticker('SBER')
resample(sber.tradestats(start="2025-01-01", end="2025-03-31"), period='1M')
```

```python
moex = moexalgo.Ticker('MOEX')
resample(moex.obstats(start="2025-01-01", end="2025-01-31"), period='1H')
```
